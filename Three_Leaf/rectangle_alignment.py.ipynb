```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
import matplotlib.pyplot as plt
```

```python

# Loading the images
rec1 = cv2.imread('/content/rec1.png')
rec2= cv2.imread('/content/rec2.png')
rec3= cv2.imread('/content/rec3.png')
rec4= cv2.imread('/content/rec4.png')

print("Original Images of rectangle1,rectangle2,rectangle3 and rectangle4 respectively")
cv2_imshow(rec1)
cv2_imshow(rec2)
cv2_imshow(rec3)
cv2_imshow(rec4)


rec1_height,rec1_width=rec1.shape[:2]
# dividing by 2 to rotate the image around its center
rotation_matrix_rec1=cv2.getRotationMatrix2D((rec1_width/2,rec1_height/2),15,1.3)
# rotation matrix and final width and height
rotated_rec1=cv2.warpAffine(rec1,rotation_matrix_rec1,(rec1_width,rec1_height))



rec2_height,rec2_width=rec2.shape[:2]
# dividing by 2 to rotate the image around its center
rotation_matrix_rec2=cv2.getRotationMatrix2D((rec2_width/1.5,rec2_height/1.5),345,1.3)
# rotation matrix and final width and height
rotated_rec2=cv2.warpAffine(rec2,rotation_matrix_rec2,(rec2_width,rec2_height))


rec3_height,rec3_width=rec3.shape[:2]
# dividing by 1.5 to rotate the image around its center
rotation_matrix_rec3=cv2.getRotationMatrix2D((rec3_width/1.5,rec3_height/1.5),330,1)
# rotation matrix and final width and height
rotated_rec3=cv2.warpAffine(rec3,rotation_matrix_rec3,(rec3_width,rec3_height))


rec1_height,rec4_width=rec4.shape[:2]
# dividing by 2 to rotate the image around its center
rotation_matrix_rec1=cv2.getRotationMatrix2D((rec4_width/2,rec1_height/2),30,1.3)
# rotation matrix and final width and height
rotated_rec4=cv2.warpAffine(rec4,rotation_matrix_rec1,(rec4_width,rec1_height))


# Display the image with numbered rectangles
print("\n \n \n")
print("Straight Image of rectangle1, rectangle2,rectangle3 and rectagle4 respectively ")
cv2_imshow(rotated_rec1)
cv2_imshow(rotated_rec2)
cv2_imshow(rotated_rec3)
cv2_imshow(rotated_rec4)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
